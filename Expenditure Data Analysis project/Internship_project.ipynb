```python
## Rename The Column That Can be incomprehensible.
```

```python
data.rename(columns ={'U/M':'Unit_of_Measure'} ,inplace=True)
data.rename(columns ={'Sales Rep':'Sales Report'} ,inplace=True)
```

```python
## Add Unit of Measure's Full Name.
```

```python
data['Unit_of_Measure'] = data['Unit_of_Measure'].replace("EA","Each")
data['Unit_of_Measure'] = data['Unit_of_Measure'].replace("SE","Some SI Units")
data['Unit_of_Measure'] = data['Unit_of_Measure'].replace("PR","Pair")
data.head()
```

```python
## Now Change The All Column Data Type.
```

```python
data['CustKey'] = data['CustKey'].astype(int)
data['DateKey'] = pd.to_datetime(data['DateKey'])
data['Discount Amount'] = data['Discount Amount'].astype(int)
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])
data['Invoice Number'] = data['Invoice Number'].astype(int)
data['Line Number'] = data['Line Number'].astype(int)
data['List Price'] = data['List Price'].astype(int)
data['Order Number'] = data['Order Number'].astype(int)
data['Promised Delivery Date'] = pd.to_datetime(data['Promised Delivery Date'])
data['Sales Amount'] = data['Sales Amount'].astype(int)
data['Sales Margin Amount'] = data['Sales Margin Amount'].astype(int)
data['Sales Price'] = data['Sales Price'].astype(int)
data['Sales Quantity'] = data['Sales Quantity'].astype(int)
data['Sales Report'] = data['Sales Report'].astype(int)
data['Sales Amount Based on List Price'] = data['Sales Amount Based on List Price'].astype(int)
data['Sales Cost Amount'] = data['Sales Cost Amount'].astype(int)
```

```python
## Import re for Remove String Valu's In Item No.
```

```python
import re
```

```python
## Remove String Value And Change Data Type.
```

```python
data['Item Number'] = data['Item Number'].apply(lambda x: re.sub(r'\D', '', str(x)))
data['Item Number'] = data['Item Number'].astype(int)
```
